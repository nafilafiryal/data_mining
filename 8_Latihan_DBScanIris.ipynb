```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('Iris.csv')

print(df.head())
```

```python
from google.colab import files

uploaded = files.upload()
```

After uploading the file, you can then load it into a pandas DataFrame. Make sure the filename matches the one you uploaded (e.g., 'Iris.csv').

```python
import io

file_name = 'Iris.csv' # Make sure this matches the uploaded file name
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

print(df.head())
```

```python
X = df[['SepalLengthCm',
        'SepalWidthCm',
        'PetalLengthCm',
        'PetalWidthCm']]
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=5)

df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)

print(df.head())
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df['PetalLengthCm'],
    y=df['PetalWidthCm'],
    hue=df['DBSCAN_Cluster'],
    palette='viridis'
)

plt.title('Hasil DBSCAN Clustering pada Data Iris')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)

plt.show()
```
