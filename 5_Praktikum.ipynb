```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

df_chap5 = pd.read_csv('Chapter05DataSet.csv')

kolom_pilihan = ['Family', 'Hobbies', 'Political', 'Professional', 'Religious', 'Social_Club', 'Support_Group']
df_chap5 = df_chap5[kolom_pilihan]

df_chap5 = df_chap5.replace({'true': True, 'false': False}).astype(bool)

frequent_items = apriori(df_chap5, min_support=0.22, use_colnames=True)

rules_chap5 = association_rules(frequent_items, metric="confidence", min_threshold=0.5)

print("Hasil Association Rules Dataset Chapter05:")
display(rules_chap5[['antecedents', 'consequents', 'support', 'confidence']])
```

```python
df_resto = pd.read_csv('Restaurant.csv', header=None, names=["Borough", "Cuisine", "Grade"])
df_resto['Cuisine'] = df_resto['Cuisine'].str.strip()
df_resto['Grade'] = df_resto['Grade'].str.strip()
df_resto = df_resto[['Cuisine', 'Grade']]

df_resto_bool = pd.get_dummies(df_resto).astype(bool)

frequent_items_resto = apriori(df_resto_bool, min_support=0.005, use_colnames=True)

rules_resto = association_rules(frequent_items_resto, metric="confidence", min_threshold=0.5)

hasil_grade_a = rules_resto[rules_resto['consequents'] == frozenset({'Grade_A'})]

print("Hasil Association Rules Dataset Restaurant (Fokus Grade A):")
display(hasil_grade_a[['antecedents', 'consequents', 'support', 'confidence']])
```
